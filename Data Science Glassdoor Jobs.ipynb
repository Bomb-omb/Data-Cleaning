```python
df.drop_duplicates(inplace=True)
df.shape
```

```python
df.reset_index(inplace=True)
df.shape
df.head()
```

```python
df.drop('index',axis=1,inplace=True)
df.head()
```

Removing numbers from company name

```python
df['Company Name'] = df['Company Name'].str.replace('\d+', '')
df['Company Name'] = df['Company Name'].str.replace('\n', '')
df['Company Name'] = df['Company Name'].str.replace('.', '')
df.head()
```

```python
df['Salary Estimate'] = df['Salary Estimate'].str.replace('K', '')
df['Salary Estimate'] = df['Salary Estimate'].str.replace('$', '')
df['Salary Estimate'] = df['Salary Estimate'].str.replace('(Glassdoor est.)', '')
df['Salary Estimate'] = df['Salary Estimate'].str.replace('(', '')
df['Salary Estimate'] = df['Salary Estimate'].str.replace(')', '')
df['Salary Estimate'] = df['Salary Estimate'].str.replace('Employer est.', '')
df.head()
```

Making a new "Job State" column with full names of states abbreviated to their respective state abbreviations and deleting 'remote' location rows

```python
df['Location'].apply(lambda x: x.split(",")[-1]).value_counts()
```

```python
df['Job State'] = df['Location'].apply(lambda x: x.split(",")[-1].strip())
df['Job State'].replace(['United States','Texas','California','New Jersey','Remote','Utah'],["US","TX","CA","NJ",np.nan,np.nan],inplace=True)
```

```python
df.dropna(axis=0,inplace=True)
df.reset_index(inplace=True)
```

Changing ratings with -1 value to 0

```python
df.Rating = np.where(df.Rating == -1.0,0, df.Rating)
df.head()
```

```python
df['Min Salary'] = df["Salary Estimate"].apply(lambda x: x.split('-')[0].strip())
df['Max Salary'] = df["Salary Estimate"].apply(lambda x: x.split('-')[-1].strip())
```

```python
df.head()
```

```python
df['Min Salary'] = pd.to_numeric(df['Min Salary'])
df['Max Salary'] = pd.to_numeric(df['Max Salary'])
df["Avg Salary"] = (df['Max Salary']+df['Min Salary'])/2
```

```python
df["Avg Salary"] = df['Avg Salary'].astype(int)
df.head()
```

Simplify Job Titles

```python
def simple_job(title):
    if 'data scientist' in title.lower():
        return 'Data Scientist'
    elif 'data engineer' in title.lower():
        return 'Data Engineer'
    elif 'analyst' in title.lower():
        return 'Analyst'
    elif 'machine learning' in title.lower():
        return 'MLE'
    elif 'manager' in title.lower():
        return 'Manager'
    elif 'director' in title.lower():
        return 'Director'
    elif 'vice president' in title.lower():
        return 'VP'
    else:
        return 'na'
df['Simple Title']=df['Job Title'].apply(simple_job)
```

```python
df.drop(['index','Founded', 'Competitors'], axis = 1, inplace=True)
```

```python
df.head()
```

Cleaning job description column and extracting key skills

```python
df['SQL'] = df['Job Description'].apply(lambda x: 1 if 'sql' in x.lower() else 0)
df['Python'] = df['Job Description'].apply(lambda x: 1 if 'python' in x.lower() else 0)
df['Excel'] = df['Job Description'].apply(lambda x: 1 if 'excel' in x.lower() else 0)
df['Hadoop'] = df['Job Description'].apply(lambda x: 1 if 'Hadoop' in x.lower() else 0)
df['AWS'] = df['Job Description'].apply(lambda x: 1 if 'aws' in x.lower() else 0)
df['Tableau'] = df['Job Description'].apply(lambda x: 1 if 'tableau' in x.lower() else 0)
df['Power BI'] = df['Job Description'].apply(lambda x: 1 if 'power bi' in x.lower() else 0)
```
